```python
import time
import json
import base64
import cv2 # type: ignore
from PIL import Image # type: ignore
from rabbit_wrapper import RabbitWrapper

from constants import RABBIT_IN_ADDR, RABBIT_IN_QNAME, RABBIT_IN_EXCH, RABBIT_IN_RKEY
#from constants import RABBIT_OUT_ADDR, RABBIT_OUT_QNAME, RABBIT_OUT_EXCH, RABBIT_OUT_RKEY


#rw = RabbitWrapper(tp='out')
rw = RabbitWrapper(tp='out', rabt_addr=RABBIT_IN_ADDR, exch_nm=RABBIT_IN_EXCH, q_nm=RABBIT_IN_QNAME, r_key=RABBIT_IN_RKEY)

image_paths = ['../imgs/1.jpg', '../imgs/2.jpg', '../imgs/3.jpg']
i = 1
for image_path in image_paths[:1]:
    with open(image_path, 'rb') as f:
        img_id = '%.20d' % i
        img_body = f.read()

        # Old implementation
        #rw.publish(bytes(img_id, "utf-8") + img_body)

        img_base64_str = base64.b64encode(img_body).decode('ascii')
        print(type(img_id), type(img_base64_str))
        jsn = {'img_id': img_id, 'img_base64': img_base64_str}
        json_str = json.dumps(jsn)  #,indent=4)
        rw.publish(json_str)
        i += 1
    time.sleep(2)
```
